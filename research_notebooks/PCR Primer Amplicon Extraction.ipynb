## Goal of Notebook: Calculate Amplicon for Two Primers

```python
# import libraries
import requests 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
### USER CONFIG VARS
target_phage = "Blessica"
# primer_forward = "ATGGTTGCCTACGGACTGTC"
# primer_reverse = "CGAACGACTTGAAGGCTTTC"

primer_forward = "GATGGTTGCCTACGGACTGT"
primer_reverse = "GAACGACTTGAAGGCTTTCG"
```

## Fasta

```python
# download the fasta file
url = "https://phagesdb.org/media/fastas/{}.fasta".format(target_phage)
r = requests.get(url, allow_redirects=True)
open('fasta_files/{}'.format(target_phage), 'wb').write(r.content) # delete the first line
```

```python
f = open('fasta_files/{}'.format(target_phage), "r")
DNA = ''.join(f.read().split("\n")[1:])
DNA
```

```python
DNA.find(primer_forward)
```

```python
DNA.find(reverse(complementation(primer_reverse)))
```

```python
comp_table = {"A":"T", "T":"A", "C":"G", "G":"C"}

def complementation(seq):
    comp_seq = ""
    for bp in seq:
        comp_seq += comp_table[bp]
    return comp_seq
    
def reverse(seq):
    return seq[::-1]
```

```python
print("Phage {} Primers:".format(target_phage))

# find primer F location
loc_F_start = DNA.find(primer_forward)+1
loc_F_stop = loc_F_start + len(primer_forward)

print("Primer Foward 5'->3': start bp {}, stop bp {}, primer length {}, primer SEQ {}".format(loc_F_start, loc_F_stop, len(primer_forward), primer_forward))

# find primer R location
loc_R_stop = DNA.find(reverse(complementation(primer_reverse)))+1
loc_R_start = loc_R_stop + len(primer_reverse)

print("Primer Reverse 3'->5': start bp {}, stop bp {}, primer length {}, primer SEQ {}".format(loc_R_start, loc_R_stop, len(primer_reverse), primer_forward))

amplicon = DNA[loc_F_stop-1:loc_R_stop-1]
print("Amplicon: start bp {}, stop bp {}, SEQ length {}, SEQ {}".format(loc_F_stop, loc_R_stop, len(amplicon), amplicon))
```
